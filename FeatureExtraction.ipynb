```python
from bnltk.pos_tagger import PosTagger  
p_tagger = PosTagger()  
p_tagger.loader()  
sentences = 'আমার নাম রাজদীপ।' 
print(p_tagger.tagger(sentences))
```

```python
import pandas as pd
```

```python
df=pd.read_csv("/content/train.csv")
df.head(20)
```

```python
df.info()
```

```python
words=df["word"].tolist()
type(words)
```

```python
len(words)
```

```python
entities=df["entity"].tolist()
type(entities)
```

```python
len(entities)
```

```python
sentence=""
number=1
mainset=[]
itl=[]
for i in words :
  if(i=="end"):
    mainset.append((number,sentence,itl,p_tagger.tagger(sentence)))
    number=number+1

    sentence=""
    itl=[]
  else :
    sentence=sentence+" "+i
    too=words.index(i)
 
  
    itl.append(entities[too])

  
```

```python
finalData = pd.DataFrame()
dataArray=[]
for data in mainset:
  sentence_number=data[0]
  sentence=data[1]
  
  splited_sen=sentence.split()
  word_count=0
  for word in splited_sen:
    first_word=False
    last_word= False
    if word_count==0:
      first_word=True
    else : first_word=False


    if (word_count==len(splited_sen)-1):
      last_word=True
    else :last_word=False

    if(first_word):
      prev_word=""
      prev_word_pos=""
    else :
      prev_word=splited_sen[word_count-1]
      prev_word_pos=data[3][word_count-1][1]

    if(last_word):
      next_word=""
      next_word_pos=""
    else :
      next_word=splited_sen[word_count+1]
      next_word_pos=data[3][word_count+1][1]


    worddata={
        "sentence_number":sentence_number,
        "sentence" : sentence,
        "word" : word,
        "entity":data[2][word_count],
        "pos":data[3][word_count][1],
        "first_word":first_word,
        "last_word":last_word,
        "prev_word":prev_word,
        "next_word":next_word,
        "prev_word_pos":prev_word_pos,
        "next_word_pos":next_word_pos
        
    }
    word_count=word_count+1;
    
    finalData=finalData.append(worddata, ignore_index=True)
    dataArray.append(worddata)





finalData.head(30)


```
